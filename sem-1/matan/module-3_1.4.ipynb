```python
# https://apps.skillfactory.ru/learning/course/course-v1:SkillFactory+MIFIML-1sem+2024/block-v1:SkillFactory+MIFIML-1sem+2024+type@sequential+block@89171c63541f45cbaa8e4dd034f010b6/block-v1:SkillFactory+MIFIML-1sem+2024+type@vertical+block@f83c474e15c146a880e9cf4fca39a262

import sympy
x, y, w = sympy.symbols('x y w')

f_origial = x**2 + 2*y**2
#  :     .  x+y=20
f = -f_origial + w*(x+y-20)
dx = f.diff(x)
dy = f.diff(y)
dw = f.diff(w)
sols = sympy.solve([dx, dy, dw], x, y, w)
print(sols)
print(f_origial.evalf(subs={'x': round(sols[x]), 'y': round(sols[y])}))
```
