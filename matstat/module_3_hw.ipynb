Для решения задачи с использованием критерия хи-квадрат и библиотеки SciPy, нам потребуется выполнить следующие шаги:

Определить ожидаемые частоты для каждого количества жалоб, исходя из предположения, что количество жалоб подчиняется распределению Пуассона со средним значением одна жалоба в месяц.

Вычислить статистику хи-квадрат, сравнив наблюдаемые и ожидаемые частоты.

Определить p-значение статистики хи-квадрат.

Сделать вывод о гипотезе на основе p-значения и уровня значимости 0.05.

Давайте начнём с определения ожидаемых частот. Мы знаем, что среднее значение распределения Пуассона равно одной жалобе в месяц, поэтому ожидаемое количество продавцов с определённым количеством жалоб можно вычислить следующим образом:

Для 0 жалоб: $e^{-1} ≈ 0,3679$ (это вероятность того, что продавец не получит ни одной жалобы).
Для 1 жалобы: $1 * e^{-1} ≈ 0,3679$ (вероятность получения одной жалобы).
Для 2 жалоб: $\frac{1}{2!} * e^{-1} ≈ 0,1839$ (вероятность получения двух жалоб).
Для более чем 3 жалоб: $(1 - 0,3679 - 0,3679 - 0,1839) ≈ 0,0803$ (вероятность получить три или более жалоб).
Теперь мы можем вычислить ожидаемые частоты:

Количество жалоб	Наблюдаемая частота	Ожидаемая частота
0	275	183,95
1	150	183,95
2	50	91,975
≥ 3	?	61,025
Мы видим, что ожидаемая частота для продавцов с тремя или более жалобами составляет примерно 61,025. Однако, поскольку мы не знаем точное количество продавцов с таким количеством жалоб, мы можем использовать общее количество продавцов (500) для вычисления ожидаемой частоты для этой категории.

Теперь давайте вычислим статистику хи-квадрат:

$χ² = Σ(\frac{(O_i - E_i)^2}{E_i})$,

где $O_i$ — наблюдаемая частота, а $E_i$ — ожидаемая частота.

Подставляя значения, получаем:

$χ² ≈ \frac{(275 - 183.95)^2}{183.95} + \frac{(150 - 183.95)^2}{183.95} + \frac{(50 - 91.975)^2}{91.975} + \frac{(x - 61.025)^2}{61.025}$,

где x — неизвестная частота для категории «≥ 3».

Для завершения расчёта нам нужно знать точное значение x. Поскольку мы не можем его определить, мы будем использовать общую сумму частот (500), чтобы найти ожидаемую частоту для категории «≥ 3»:

$500 - (275 + 150 + 50) = 125$.

Это даёт нам возможность рассчитать статистику хи-квадрат без учёта категории «≥ 3», так как её вклад будет минимальным по сравнению с другими категориями.

Используя Python и библиотеку SciPy, мы можем выполнить эти расчёты и сравнить результаты с уровнем значимости 0.05, чтобы сделать вывод о нашей гипотезе.

Обратите внимание, что этот подход предполагает, что распределение Пуассона является подходящим для описания количества жалоб на продавцов. В реальности могут быть другие факторы, влияющие на распределение жалоб, которые следует учитывать при анализе данных.

Для решения задачи проверки гипотезы о распределении Пуассона с использованием критерия хи-квадрат воспользуемся библиотекой `scipy` для Python. Вот пошаговый план:

### Шаг 2: Определим данные

У нас есть следующие данные:
- Всего продавцов: 500
- Продавцов без жалоб: 275
- Продавцов с одной жалобой: 150
- Продавцов с двумя жалобами: 50
- Остальные продавцы имеют 3 или более жалоб.

Сначала найдем количество продавцов с тремя и более жалобами:

$$ \text{Продавцы с 3+ жалобами} = 500 - 275 - 150 - 50 = 25 $$

Теперь у нас есть все данные:
- $n_0 = 275$ (продавцов без жалоб)
- $n_1 = 150$ (с одной жалобой)
- $n_2 = 50$ (с двумя жалобами)
- $n_{3+} = 25$ (с тремя и более жалобами).

### Шаг 3: Проверка гипотезы о распределении Пуассона

Гипотеза состоит в том, что количество жалоб на одного продавца следует распределению Пуассона с параметром $\lambda = 1$. Для этого мы должны вычислить теоретические вероятности для каждой категории и сравнить их с наблюдаемыми данными.

#### Теоретическое распределение Пуассона

Распределение Пуассона задается формулой:

$$ P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!} $$

где $\lambda$ — среднее значение, а $e$ — основание натурального логарифма.

Вычислим вероятности для значений $X = 0$, $X = 1$, $X = 2$, и $X \geq 3$.

#### Вероятность того, что нет жалоб ($P(X=0)$):

$$ P(X = 0) = \frac{e^{-1}}{0!} = e^{-1} \approx 0.3679 $$

#### Вероятность того, что одна жалоба ($P(X=1)$):

$$ P(X = 1) = \frac{1^1 e^{-1}}{1!} = e^{-1} \approx 0.3679 $$

#### Вероятность того, что две жалобы ($P(X=2)$):

$$ P(X = 2) = \frac{1^2 e^{-1}}{2!} = \frac{e^{-1}}{2} \approx 0.1839 $$

#### Вероятность того, что три или более жалоб ($P(X \geq 3)$):

$$ P(X \geq 3) = 1 - P(X = 0) - P(X = 1) - P(X = 2) = 1 - 0.3679 - 0.3679 - 0.1839 = 0.0803 $$

### Шаг 4: Вычисление ожидаемых частот

Ожидаемые частоты рассчитываются умножением общей численности продавцов на соответствующие вероятности:

- Ожидаемое число продавцов без жалоб: $E_0 = 500 \times 0.3679 \approx 184$
- Ожидаемое число продавцов с одной жалобой: $E_1 = 500 \times 0.3679 \approx 184$
- Ожидаемое число продавцов с двумя жалобами: $E_2 = 500 \times 0.1839 \approx 92$
- Ожидаемое число продавцов с тремя и более жалобами: $E_{3+} = 500 \times 0.0803 \approx 40$

### Шаг 5: Расчет статистики хи-квадрат

Статистика хи-квадрат рассчитывается по формуле:

$$ X^2 = \sum \frac{(O_i - E_i)^2}{E_i} $$

где $O_i$ — наблюдаемые значения, а $E_i$ — ожидаемые значения.

Подставляем наши данные:

$$ X^2 = \frac{(275 - 184)^2}{184} + \frac{(150 - 184)^2}{184} + \frac{(50 - 92)^2}{92} + \frac{(25 - 40)^2}{40} $$

Выполним расчет:

$$ X^2 = \frac{(91)^2}{184} + \frac{(-34)^2}{184} + \frac{(-42)^2}{92} + \frac{(-15)^2}{40} $$

$$ X^2 = \frac{8281}{184} + \frac{1156}{184} + \frac{1764}{92} + \frac{225}{40} $$

$$ X^2 = 44.96 + 6.28 + 19.17 + 5.63 = 76.04 $$

### Шаг 6: Определение критического значения и принятие решения

Для уровня значимости $\alpha = 0.05$ и числа степеней свободы $df = 3$ (так как мы объединяем категорию "три и более"), критическое значение $\chi^2$ равно приблизительно 7.815.

Так как полученное нами значение $X^2 = 76.04$ значительно превышает критическое значение, мы отвергаем нулевую гипотезу о том, что количество жалоб на продавца подчиняется распределению Пуассона с параметром $\lambda = 1$.

````python
n0 = np.exp(-1) * 500
n1 = 1 * np.exp(-1) * 500
n2 = (1 / 2) * np.exp(-1) * 500
n3 = 500 - n0 - n1 - n2

chi2_stat = (275 - n0)**2/n0 + (150-n1)**2/n1 + (50 - n2)**2/n2 + (25 - n3)**2/n3
round(chi2_stat, 2)
````
